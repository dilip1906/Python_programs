### Can you find how many deliveries were faced by batsman  `SC Ganguly`.

```python
y = data.get('innings')

a = y[0]

b = a.get('1st innings').get('deliveries')


counter = 0

for i in range(len(b)):
  for key in b[i].keys():
    if b[i].get(key).get('batsman')=='SC Ganguly':
      counter+=1
print('SC Ganguly faced {} deliveries '.format(counter))
```

### Who was man of the match and how many runs did he scored ?

```python
print('The player of the match is: ', data.get('info').get('player_of_match')[0])

y = data.get('innings')

a = y[0]

b = a.get('1st innings').get('deliveries')

runs = []

for i in range(len(b)):
  for key in b[i].keys():
    if b[i].get(key).get('batsman')=='BB McCullum':
      run = b[i].get(key).get('runs').get('batsman')
      runs.append(run)
print('BB McCullum scored {} runs'.format(sum(runs)))
```

### Which batsman played in the first inning?

```python
bat = []

for i in range(len(b)):
  for key in b[i].keys():
    bat_man = b[i].get(key).get('batsman')
    bat.append(bat_man)


res = []

for i in bat:
  if i not in res:
    res.append(i)
    
print('The list of batsmen who played in the first innings are: ',res)
```
